```python
import numpy as np
import matplotlib.pyplot as plt

time_index = 1000
# Define the size of the matrix
n = 5  # You can change this to any desired size

# Create an n x n matrix with random values
matrix = np.array([[0.24, 0.73], [0.01, 0.31]])

# Plot the heatmap
plt.imshow(matrix, cmap='coolwarm', interpolation='nearest', vmin=-1, vmax=1)
plt.colorbar()
plt.title(f'{n} x {n} estimated B in time-step {time_index}')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()

matrix = np.array([[0, 1.2], [0, 0]])

# Plot the heatmap
plt.imshow(matrix, cmap='coolwarm', interpolation='nearest', vmin=-1, vmax=1)
plt.colorbar()
plt.title(f'{n} x {n} B gt in time-step {time_index}')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```
