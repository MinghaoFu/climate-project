```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.array([1, 2, 3, 4])

# Calculate y values for y = x
y_linear = x

# Calculate y values for y = x + (x-1)(x-2)(x-3)*(x-4)
y_polynomial = x + (x-1)*(x-2)*(x-3)*(x-4)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y_linear, label='y = x', marker='o')
plt.plot(x, y_polynomial, label='y = x + (x-1)(x-2)(x-3)*(x-4)', marker='o')

plt.title('Plot of y = x and y = x + (x-1)(x-2)(x-3)*(x-4)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
